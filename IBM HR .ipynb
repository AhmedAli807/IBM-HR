```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import sqlite3 as sql
```

```python
df=pd.read_excel('F:/data analysis/IBM HR proj/IBM HR.xlsx')
df.head()
```

```python
df.info()
df.describe()
```

```python
df['Attrition'].value_counts(normalize='True')
```

```python
conn=sql.connect(':memory:')
df.to_sql('hr', conn, index=False)

query = """
SELECT Department,
       COUNT(*) AS TotalEmployees,
       SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END) AS Resigned,
       ROUND(100.0 * SUM(CASE WHEN Attrition = 'Yes' THEN 1 ELSE 0 END)/COUNT(*), 2) AS AttritionRate
FROM hr
GROUP BY Department
ORDER BY AttritionRate DESC
"""

pd.read_sql(query, conn)
```

```python
attrition_by_dept = pd.read_sql(query, conn)
sns.barplot(data=attrition_by_dept, x='AttritionRate', y='Department')
plt.title("Attrition By Department")
plt.xlabel("Attrition(%)")
plt.ylabel("Department")
plt.show()
```

```python
avg_emp_query="""

    SELECT JobLevel, ROUND(AVG(MonthlyIncome), 2) AS AvgSalary
    FROM hr
    GROUP BY JobLevel
"""
job_level_by_salary=pd.read_sql(avg_emp_query,conn)
sns.barplot(data=job_level_by_salary,x="JobLevel",y="AvgSalary")
plt.title("Job Level By Salary")
plt.xlabel("Job Level")
plt.ylabel("Avg Salary")
```

```python
perf_by_attr_query="""
SELECT PerformanceRating, Attrition, COUNT(*) AS Count
    FROM hr
    GROUP BY PerformanceRating, Attrition
"""
perf_by_attr=pd.read_sql(perf_by_attr_query,conn)
plt.figure(figsize=(8,5))
sns.barplot(data=perf_by_attr, x='PerformanceRating', y='Count', hue='Attrition')
plt.title('Attrition by Performance Rating')
plt.xlabel('Performance Rating')
plt.ylabel('Employee Count')
plt.show()
```

```python
attr_by_years_query="""
    SELECT YearsAtCompany, Attrition, COUNT(*) AS Count
    FROM hr
    GROUP BY YearsAtCompany, Attrition
"""
attr_by_years=pd.read_sql(attr_by_years_query,conn)
plt.figure(figsize=(10,6))
sns.lineplot(data=attr_by_years, x='YearsAtCompany', y='Count', hue='Attrition', marker='o')
plt.title('Attrition by Years at Company')
plt.xlabel('Years at Company')
plt.ylabel('Number of Employees')
plt.show()
```
